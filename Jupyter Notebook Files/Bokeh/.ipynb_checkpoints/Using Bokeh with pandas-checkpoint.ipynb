```python
import pandas
from bokeh.plotting import figure, output_file, show
     
p = figure(plot_width=500,plot_height=400, tools='pan')
     
p.title.text="Cool Data"
p.title.text_color="Gray"
p.title.text_font="times"
p.title.text_font_style="bold"
p.xaxis.minor_tick_line_color=None
p.yaxis.minor_tick_line_color=None
p.xaxis.axis_label="Date"
p.yaxis.axis_label="Intensity"    
     
p.line([1,2,3],[4,5,6])
output_file("graph.html")
show(p)

#You can add a title to the plot, set the figure width and height, change title font, etc. 
#Above is a summary of properties which can be added to change the style of the plo.
```

```python
import pandas
from bokeh.plotting import figure, output_file, show

output_file("TempPressureCirclePlot.html")

df = pandas.read_excel("http://pythonhow.com/data/verlegenhuken.xlsx")

df["Temperature"] = df["Temperature"]/10
df["Pressure"] = df["Pressure"]/10

f = figure()

f.title.text = ("Temperature and Air Pressure")
f.yaxis.axis_label = "Pressure(hPa)"
f.xaxis.axis_label = "Temperature(oC)"

f.circle(df["Temperature"], df["Pressure"], size = 0.5)

show(f)
```

```python
# Changing visual Attributes in Bokeh

from bokeh.plotting import figure, output_file, show
p = figure(plot_width=500, plot_height=400, tools = 'pan, reset')
p.title.text = "Earthquakes"
p.title.text_color = "Orange"
p.title.text_font = "times"
p.title.text_font_style = "italic"
p.yaxis.minor_tick_line_color = "Yellow"
p.xaxis.axis_label = "Times"
p.yaxis.axis_label = "Value"
p.circle([1,2,3,4,5], [5,6,5,5,3], size = [i*2 for i in [8,12,14,15,20]], color="red", alpha=0.5)
output_file("Scatter_plotting.html")
show(p)

# alpha is a property that denotes transparency
```

```python
#How to plot time series data

from bokeh.plotting import output_file, show, figure
import pandas

df = pandas.read_csv('adobe.csv', parse_dates = ["Date"])

f = figure(height = 500, width = 1000, x_axis_type = "datetime", sizing_mode "scale_width")
f.line(df["Date"], df["Close"])

output_file("timeseriesdata.html")

show(f)
```
